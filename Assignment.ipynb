```python
import requests
from datetime import datetime

#API_KEY = "your_api_key_here"  # Replace with your actual API key


def convert_to_unix_timestamp(date_str):
    dt = datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")
    timestamp = int(dt.timestamp())
    return timestamp

def get_weather_data(response_data):
    date_str = input("Enter the date (YYYY-MM-DD): ")
    date_unix_timestamp = convert_to_unix_timestamp(date_str)

    if "list" in response_data:
        # Find the entry in the 'list' that matches the input date
        weather_data = next(
            (entry for entry in response_data["list"] if entry["dt"] == date_unix_timestamp),
            None
        )

        if weather_data:
            temperature = weather_data["main"]["temp"]
            print(f"Weather report for {date_str}: {temperature} K")
        else:
            print(f"No weather data available for {date_str}.")
    else:
        print("Invalid API response format. 'list' key not found.")




def get_wind_speed_data(response_data):
    date_str = input("Enter the date (YYYY-MM-DD): ")
    date_unix_timestamp = convert_to_unix_timestamp(date_str)

    # Find the entry in the 'list' that matches the input date
    weather_data = next(
        (entry for entry in response_data["list"] if entry["dt"] == date_unix_timestamp),
        None
    )

    if weather_data:
        wind_speed = weather_data["wind"]["speed"]
        print(f"Wind speed for {date_str}: {wind_speed} m/s")
    else:
        print(f"No wind speed data available for {date_str}.")

def get_pressure_data(response_data):
    date_str = input("Enter the date (YYYY-MM-DD): ")
    date_unix_timestamp = convert_to_unix_timestamp(date_str)

    # Find the entry in the 'list' that matches the input date
    weather_data = next(
        (entry for entry in response_data["list"] if entry["dt"] == date_unix_timestamp),
        None
    )

    if weather_data:
        pressure = weather_data["main"]["pressure"]
        print(f"Pressure for {date_str}: {pressure} hPa")
    else:
        print(f"No pressure data available for {date_str}.")

def main():
    while True:
        print("\nMenu:")
        print("1. Get weather")
        print("2. Get Wind Speed")
        print("3. Get Pressure")
        print("0. Exit")

        choice = int(input("Enter your choice: "))

        if choice == 1:
            get_weather_data(response_data)
        elif choice == 2:
            get_wind_speed_data(response_data)
        elif choice == 3:
            get_pressure_data(response_data)
        elif choice == 0:
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    # Replace this line with the actual API request and JSON response
    response_data = {}  # Placeholder for the API response data
    main()
```
